```
import pandas as pd
import numpy as np
from scipy.sparse.linalg import svds

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

table = ratings.pivot("userId", "movieId", "rating").fillna(0)
table.head()
```

```
from sklearn.decomposition import NMF

x = table.values
model = NMF(50, init='random', random_state=0)
nmf_features = model.fit_transform(x)
nmf_features_components = model.components_
predictions = pd.DataFrame(nmf_features_components, columns = table.columns)
predictions
```

```
#Creates User Rating Data and Returns it
def User_Rating_Data(ratings, user, movies):
  user_rating_info = ratings[ratings.userId == user]
  user_info = user_rating_info.merge(movies, how = "left", left_on = "movieId", right_on = "movieId").sort_values(["rating"], ascending = False)
  
  return user_info

#Recommends a List of Movies to User that he/she may like
def Recommend_Movie_To_User(user_info, predictions, recommended_movie_amount, movies):
  row_user = user_info["userId"][0] - 1
  predictions_sorted = predictions.iloc[row_user].sort_values(ascending = False)
  movie_recommendations = (movies[~movies["movieId"].isin(user_info["movieId"])].merge(pd.DataFrame(predictions_sorted).reset_index(), how = "left",left_on = "movieId", right_on = "movieId").rename(columns = {row_user: "predictions"}).sort_values("predictions", ascending = False).iloc[:recommended_movie_amount, :-1])

  return movie_recommendations
```

```
#Example of a list of movies that a particular user has rated
print("Movies that were rated by the user:")
User_Number_9 = User_Rating_Data(ratings, 9, movies)
User_Number_9
```

```
#Example of a list of movies to recommend the user
print("Movies that were recommended to the user:")
User_Number_9_Recommendations = Recommend_Movie_To_User(User_Number_9, predictions, 10, movies)
User_Number_9_Recommendations
```
